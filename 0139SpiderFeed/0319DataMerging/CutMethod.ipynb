```python
# 分词表
method_list = ['木马后门','社会工程','远程控制','电子邮件','钓鱼攻击','勒索软件','漏洞利用','僵尸网络','恶意软件',
            '恶意扫描','挖矿','RAT','鱼叉式网络钓鱼攻击','伪造程序','诱饵文件','DDOS攻击','拒绝服务攻击','供应链攻击','恶意脚本',
            '文件篡改','Redis未授权命令执行','垃圾邮件','伪装文件','水坑攻击','病毒','伪造证书','网站篡改','无文件攻击']
```

```python
def cutword(word):
    jieba.load_userdict("dict.txt")
    wordList = jieba.cut(word)
    res = []
    for w in wordList:
        res.append(w)
    return res
```

```python
f = open("ByAPT1.csv", encoding='utf-8-sig')
data = pd.read_csv(f, encoding='utf-8-sig')
data["攻击方法"] = data["攻击方法"].fillna("")
data["攻击方法"] = data["攻击方法"].apply(cutword)
```
